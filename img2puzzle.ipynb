```python
from PIL import Image as imgz
import os
import glob
```

```python
myWidth = 2
myHeight = 2

ExampleImage = imgz.open("Images/Bliss.png")


for d in glob.glob('Results/*'):
    try:
        os.remove(d)
    except:
        for f in glob.glob(d+'/*.png'):
            os.remove(f)

sizeX = round(ExampleImage.size[0]/ myWidth)
sizeY = round(ExampleImage.size[1] / myHeight)
PicLarg = round(sizeY/5)
PicSize = round(sizeY/10)

picSizeRatio = 1.5
ratioCoteVer = (round(sizeY/2 - PicSize*picSizeRatio),round(sizeY/2 + PicSize*picSizeRatio))
ratioCoteHor = (round(sizeX/2 - PicSize*picSizeRatio), round(sizeX/2 + PicSize*picSizeRatio))

for x in range (myWidth):
    for y in range(myHeight):
        
        boxiLeft = sizeX*x
        boxiUp = sizeY*y
        boxiRight = sizeX*(x+1)
        boxiDown = sizeY*(y+1)
        boxiPics = {"up": [],"left": [], "right": [],"down": []}
        # --------------------------------------------UP-------------------------------------------------------
        if (y != 0): 

            box = (sizeX*x, sizeY*y, sizeX*(x+1), sizeY*y+PicLarg)

            b1 = (sizeX*x, sizeY*y, ratioCoteHor[0]+sizeX*(x), sizeY*y+PicLarg)
            b2 = (ratioCoteHor[0]+sizeX*(x), sizeY*y, ratioCoteHor[1]+sizeX*(x), sizeY*y+PicLarg)
            b3 = (ratioCoteHor[1]+sizeX*(x), sizeY*y, sizeX*(x+1), sizeY*y+PicLarg)

            boxes = [b1,b2,b3]

            for k in range(len(boxes)):
                mb = ExampleImage.crop(boxes[k])
                boxiPics["up"].append(mb)

            boxiUp = sizeY*y +PicLarg
        else:
            pass
        # -----------------------------------------------LEFT--------------------------------------------------
        if (x != 0):
            
            box = (sizeX*x, sizeY*y, sizeX*x+PicLarg, sizeY*(y+1))

            b1 = (sizeX*x, sizeY*y, sizeX*x+PicLarg, ratioCoteVer[0]+sizeY*(y))
            b2 = (sizeX*x, ratioCoteVer[0]+sizeY*(y), sizeX*x+PicLarg, ratioCoteVer[1]+sizeY*(y))
            b3 = (sizeX*x, ratioCoteVer[1]+sizeY*(y), sizeX*x+PicLarg, sizeY*(y+1))

            boxes = [b1,b2,b3]

            for k in range(len(boxes)):
                mb = ExampleImage.crop(boxes[k])
                boxiPics["left"].append(mb)

            boxiLeft = sizeX*x + PicLarg
        else:
            pass
        # -----------------------------------------------RIGHT-------------------------------------------------
        if (x != myWidth -1):
            box = (sizeX*(x+1)-PicLarg, sizeY*y,sizeX*(x+1), sizeY*(y+1))

            b1 = (sizeX*(x+1)-PicLarg, sizeY*y, sizeX*(x+1), ratioCoteVer[0]+sizeY*(y))
            b2 = (sizeX*(x+1)-PicLarg, ratioCoteVer[0]+sizeY*(y), sizeX*(x+1), ratioCoteVer[1]+sizeY*(y))
            b3 = (sizeX*(x+1)-PicLarg, ratioCoteVer[1]+sizeY*(y), sizeX*(x+1), sizeY*(y+1))

            boxes = [b1,b2,b3]

            for k in range(len(boxes)):
                mb = ExampleImage.crop(boxes[k])
                boxiPics["right"].append(mb)

            boxiRight = sizeX*(x+1) - PicLarg
        else:
            pass        
        # --------------------------------------------------DOWN-----------------------------------------------
        if (y != myHeight -1):
            box = (sizeX*x, sizeY*(y+1)-PicLarg, sizeX*(x+1), sizeY*(y+1))

            b1 = (sizeX*x, sizeY*(y+1)-PicLarg, ratioCoteHor[0]+sizeX*(x), sizeY*(y+1))
            b2 = (ratioCoteHor[0]+sizeX*(x), sizeY*(y+1)-PicLarg, ratioCoteHor[1]+sizeX*(x), sizeY*(y+1))
            b3 = (ratioCoteHor[1]+sizeX*(x), sizeY*(y+1)-PicLarg, sizeX*(x+1), sizeY*(y+1))

            boxes = [b1,b2,b3]

            for k in range(len(boxes)):
                mb = ExampleImage.crop(boxes[k])
                boxiPics["down"].append(mb)

            boxiDown = sizeY*(y+1) -PicLarg
        else:
            pass
        # -----------------------------------------------------------------------------------------------------
        boxi = (boxiLeft,boxiUp,boxiRight,boxiDown)
        im = ExampleImage.crop(boxi)
        im = im.save("Results/"+str(x) + "," + str(y) + ".png")
        for ele in boxiPics:
            if boxiPics[ele] != []:
                if not (os.path.exists("Results/" + ele)):
                    os.makedirs("Results/" + ele)
                for k in range (len(boxiPics[ele])):
                     (boxiPics[ele][k]).save("Results/"+ele+"/("+ str(x) + "," + str(y) + ")_" +str(k)+".png")
```

```python
PieceBase = imgz.open("Results/0,0.png")
CoteDroite = [imgz.open("Results/right/(0,0)_"+ str(k)+".png") for k in range(3)]
CoteBas = [imgz.open("Results/down/(0,0)_"+ str(k)+".png") for k in range(3)]

t = (PieceBase.size[0] + CoteDroite[0].size[0],PieceBase.size[1] + CoteBas[0].size[1])
Piece = imgz.new("RGBA",t)
Piece.paste(PieceBase,(0,0))

Piece.paste(CoteDroite[0],(PieceBase.size[0],0))
#Piece.paste(CoteDroite[1],(PieceBase.size[0],CoteDroite[0].size[1]))
Piece.paste(CoteDroite[2],(PieceBase.size[0],CoteDroite[0].size[1]+CoteDroite[1].size[1]))

Piece.paste(CoteBas[0],(0,PieceBase.size[1]))
#Piece.paste(CoteBas[1],(CoteBas[0].size[0],PieceBase.size[1]))
Piece.paste(CoteBas[2],(CoteBas[0].size[0]+CoteBas[1].size[0],PieceBase.size[1]))

Piece.save("Results/resultatTest.png")
```
